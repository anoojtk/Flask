```python
df.drop(['Property_Area','Loan_Amount_Term','CoapplicantIncome','Loan_ID','Credit_History','Married','Dependents','Gender'],axis=1,inplace=True)
```

```python

df['Education']=df['Education'].replace({'Graduate':1,'Not Graduate':0})
df['Loan_Status']=df['Loan_Status'].replace({'Y':1,'N':0})
df['Self_Employed']=df['Self_Employed'].replace({'No':0,'Yes':0})




```

```python
df.isnull().sum()
```

```python
df=df.fillna(df.median())
```

```python
df
```

```python
y=df['Loan_Status']
x=df.drop(['Loan_Status'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn import linear_model
lr=linear_model.LogisticRegression()
model = lr.fit(x_train,y_train)
pickle.dump(lr,open('model.pkl','wb'))
```

```python
pre=model.predict(x_test)
pre
```

```python
print(pre)
```

```python
from sklearn.metrics import mean_squared_error
print('mse error is :',mean_squared_error(y_test,pre))
```
